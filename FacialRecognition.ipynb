```python
# üì¶ Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay

from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier

from imblearn.over_sampling import SMOTE

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# üì• Load Dataset
df = pd.read_csv("C:/Users/shaik/OneDrive/Desktop/FacialRecognitionProject/FacialRecognitionProject.ipynb/facerec.csv")

# üéØ Split Features & Labels
features = df.drop(columns=['Label']).values
labels = df['Label'].values

# ‚öôÔ∏è Train-Test Split with Stratification
X_train, X_test, y_train, y_test = train_test_split(
    features, labels, test_size=0.2, stratify=labels, random_state=42
)

# üîÑ Balance Training Data using SMOTE
sm = SMOTE(random_state=42)
X_train_bal, y_train_bal = sm.fit_resample(X_train, y_train)

# üßΩ Feature Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_bal)
X_test_scaled = scaler.transform(X_test)

# üõ†Ô∏è Define Model Dictionary
model_dict = {
    "Random Forest": RandomForestClassifier(random_state=42),
    "XGBoost": XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='mlogloss'),
    "SVM": SVC(probability=True, random_state=42)
}

# üîß Hyperparameter Grids
search_grids = {
    "Random Forest": {'n_estimators': [100, 300, 500], 'max_depth': [10, 15, 20]},
    "XGBoost": {'n_estimators': [100, 300, 500], 'max_depth': [6, 10, 15]},
    "SVM": {'C': [0.1, 1, 10], 'gamma': ['scale', 0.1, 0.01]}
}

# üìà Model Results Container
model_results = {}

# üîÅ Train Each Traditional Model
for name, clf in model_dict.items():
    print(f'\nüîç Training {name}...')
    search = RandomizedSearchCV(
        clf, search_grids[name], n_iter=5, cv=3, scoring='accuracy',
        random_state=42, verbose=1, n_jobs=-1
    )
    search.fit(X_train_scaled, y_train_bal)
    best_clf = search.best_estimator_

    predictions = best_clf.predict(X_test_scaled)
    acc = accuracy_score(y_test, predictions)
    model_results[name] = acc * 100

    print(f'‚úÖ {name} Accuracy: {acc * 100:.2f}%')
    print('üìã Classification Report:')
    print(classification_report(y_test, predictions))

    cm = confusion_matrix(y_test, predictions)
    ConfusionMatrixDisplay(confusion_matrix=cm).plot(cmap='Blues')
    plt.title(f'{name} Confusion Matrix')
    plt.show()

# üß† CNN Training
print('\nüß† Training CNN Model...')

# üîÅ Reshape for CNN Input
X_train_cnn = X_train_scaled.reshape(-1, 15, 10, 1)
X_test_cnn = X_test_scaled.reshape(-1, 15, 10, 1)

le = LabelEncoder()
y_train_cnn = le.fit_transform(y_train_bal)
y_test_cnn = le.transform(y_test)
n_classes = len(np.unique(labels))

# üß± CNN Architecture
cnn = Sequential([
    Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(15, 10, 1)),
    MaxPooling2D(pool_size=(2, 2)),
    Dropout(0.25),

    Conv2D(64, kernel_size=(3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Dropout(0.25),

    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(n_classes, activation='softmax')
])

# üõ†Ô∏è Compile & Train CNN
cnn.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

cnn.fit(
    X_train_cnn, y_train_cnn,
    epochs=50,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stopping],
    verbose=1
)

# üìä Evaluate CNN
cnn_eval = cnn.evaluate(X_test_cnn, y_test_cnn, verbose=0)
model_results['CNN'] = cnn_eval[1] * 100
print(f'‚úÖ CNN Accuracy: {cnn_eval[1] * 100:.2f}%')

# üìä Plot Model Accuracy Comparison
plt.figure(figsize=(10, 6))
sns.barplot(x=list(model_results.keys()), y=list(model_results.values()), palette='Set2')
plt.title('üîé Model Accuracy Comparison', fontsize=16)
plt.ylabel('Accuracy (%)')
plt.xlabel('Model')
plt.ylim(0, 100)
plt.grid(axis='y')
plt.show()

# üìå Print Final Accuracy Summary
print('\nüìå Final Accuracy Results:')
for model, score in model_results.items():
    print(f'{model}: {score:.2f}%')
```
